# Find the Most Expensive Apps, Filter out the Junk, and Calculate a (ballpark) Sales Revenue Estimate

Let's examine the Price column more closely.

**Challenge**: Convert the price column to numeric data. Then investigate the top 20 most expensive apps in the dataset.

Remove all apps that cost more than $250 from the `df_apps_clean` DataFrame.

Add a column called 'Revenue_Estimate' to the DataFrame. This column should hold the price of the app times the number of installs. What are the top 10 highest grossing paid apps according to this estimate? Out of the top 10 highest grossing paid apps, how many are games?


```python
# Price is a object type because of the $ sign.
df_apps_clean["Price"].describe()

# Removing the $ symbol by converting the object to a string and removing the $ after.

df_apps_clean["Price"] = df_apps_clean["Price"].astype(str).str.replace('$', "")

# Convert to numeric

df_apps_clean["Price"] = pd.to_numeric(df_apps_clean["Price"])

df_apps_clean.sort_values(by="Price", ascending=False).head(20)
```

```python
# Removing rows that are greater than $250.

df_apps_clean = df_apps_clean[df_apps_clean["Price"] < 250]

# Checking the top 20 most expensive apps again. I am rich is still in there...
# But I shall not remove outliers as that introduces bias...
```

# Grouped Bar Charts: Free vs. Paid Apps per Category

```python
# Make a new dataframe of paid vs free apps

df_free_vs_paid = df_apps_clean.groupby(["Category", "Type"], as_index=False).agg({"App": pd.Series.count})
df_free_vs_paid

# Sort by number of apps

df_free_vs_paid_sorted = df_free_vs_paid.sort_values("App", ascending=False)
df_free_vs_paid_sorted
```

```python
# Creating a bar chart of the number of apps based of free vs paid by each category.
# Since I already sorted the data into a new dateframe, I won't be using the total desending method.
fig_7 = px.bar(df_free_vs_paid_sorted,
               x= "Category",
               y="App",
               color="Type",
               barmode="group", # Creates a side by side bar chart.
               title="Free vs Paid Apps by Category")

# Updating the graph with a log scale and y axis title. X axis label is fine how it is.
# I also centered the title.

fig_7.update_layout(yaxis=dict(type="log"),
                    yaxis_title="Number of Apps",
                    title_x=0.5)
fig_7.show()
```
