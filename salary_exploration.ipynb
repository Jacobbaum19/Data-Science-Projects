```python
# Importing the files
## Note to self: Cntrl + space to autofile

import pandas as pd

df = pd.read_csv("salaries_by_college_major.csv")
df.head(5)

```

```python
# Checking the shape
df.shape

# Checking the column names
df.columns

# Check for NA values
df.isna()

# Clearing the last row

clean_df = df.dropna()
clean_df.tail(5)
```

```python

# Accessing the Starting Median Salary Column

clean_df['Starting Median Salary'].max()

# Seeing which job makes $74,300 as a starting salary!

clean_df['Starting Median Salary'].idxmax()

# Seeing row 43
# Two ways of doing this, one with the specfic column
# And one with all the other infomation.
clean_df["Undergraduate Major"].iloc[43]
clean_df.iloc[43]

```

```python
# Exercises

# 1) Highest mid-carerr salary

clean_df["Mid-Career Median Salary"].idxmax()

# Finding the 8th row
clean_df.iloc[8]

# Answer: Chem Eng. with $107,000 Mid-Career Salary
```

```python
# 2) Lowest starting salary
clean_df["Starting Median Salary"].idxmin()

# Checking row 49
clean_df.iloc[49]

# Answer: Spanish
```

```python
# 3) Lowest Mid-Career salary

clean_df["Mid-Career Median Salary"].idxmin()

# Checking row 18
clean_df.iloc[18]

# Answer: Education, $52,000
```

```python
# Adding a new column called Risk

# Commented it out becase it exists in the local memory
# spread_col = clean_df["Mid-Career 90th Percentile Salary"] - clean_df["Mid-Career 10th Percentile Salary"]
# clean_df.insert(1, "Spread", spread_col)
print(clean_df.head(5))
```

```python
# Seeing the low-risk majors

low_risk = clean_df.sort_values("Spread", ascending=True)
low_risk[["Undergraduate Major", "Spread"]].head(5)
```

```python
# Finding the top 5 highest values in 90th percentile. Also check the largest spread.

highest_10th_percentile = clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending=False)
print(highest_10th_percentile[["Undergraduate Major", "Mid-Career 90th Percentile Salary"]].head(5))

# Answer: Economics, Finance, Chem. Eng., Math, Physics.

# Finding the greatest spread

great_spread = clean_df.sort_values("Spread", ascending=False)
great_spread[["Undergraduate Major", "Spread"]].head(5)

# Answer: Economics, Finance, Math, Marketing and Philosophy
```

```python
# Group by function...
clean_df.groupby('Group').count()
clean_df.groupby("Group").mean(numeric_only=True)

# Try again with the $ symbol.
pd.options.display.float_format = "{:,.2f}".format
clean_df.groupby("Group").mean(numeric_only=True)
```
